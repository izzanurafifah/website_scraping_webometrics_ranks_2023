```python
from bs4 import BeautifulSoup
import requests
```

```python
base_url = 'https://www.webometrics.info/en/world?page={}'

urls = []
pages = 10

for page in range(pages):
  url = base_url.format(page)
  urls.append(url)

for url in urls:
  page = requests.get(url)
  soup = BeautifulSoup(page.text)

header = [th.text for th in soup.find_all('th')]

header
```

```python
universities = []

for url in urls:
  page = requests.get(url)
  soup = BeautifulSoup(page.text)
  table = [data.text for data in soup.find_all('td')]
  universities.extend(table)


print(universities)
```

```python
import pandas as pd

ranking = universities[0::7]
university = universities[1::7]
det = universities[2::7]
country = universities[3::7]
impact_rank = universities[4::7]
openness_rank = universities[5::7]
excellence_rank = universities[6::7]
```

```python
df = pd.DataFrame({
    header[0]: ranking,
    header[1]: university,
    header[2]: det,
    header[3]: country,
    header[4]: impact_rank,
    header[5]: openness_rank,
    header[6]: excellence_rank
})

print(df)
```
